```python
def kdeplots(df):
    columns = [column for column in df.columns if df[column].max()<=1]

    df = df.drop(columns,axis=1)
    columnCounts = len(df.columns)
    fig,axes = plt.subplots(1,columnCounts,figsize=(64,8))

    for column in df.columns:
        columnCounts-=1
        sns.kdeplot(df[column],ax=axes[columnCounts])
```
